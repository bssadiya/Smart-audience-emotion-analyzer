```python
import cv2
import os
import random
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
import shutil
import uuid

# =========================
# CONFIGURATION
# =========================
input_root = "images/dataset/train"                   # original dataset
output_root = "images/dataset/train_cleaned"          # cleaned + augmented images
target_size = (128, 128)

import shutil

def clear_folder(path):
    if os.path.exists(path):
        shutil.rmtree(path)
    os.makedirs(path, exist_ok=True)

# Example before preprocessing
clear_folder(output_root)



# Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# =========================
# CLAHE
# =========================
def apply_clahe(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced = clahe.apply(gray)
    return cv2.cvtColor(enhanced, cv2.COLOR_GRAY2BGR)

# =========================
# Data Augmentation
# =========================
def augment_image(img):
    augmented = []

    # Horizontal flip
    augmented.append(cv2.flip(img, 1))

    # Small rotations
    for angle in [-15, -10, 10, 15]:
        rows, cols, _ = img.shape
        M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
        rotated = cv2.warpAffine(img, M, (cols, rows), borderMode=cv2.BORDER_REFLECT)
        augmented.append(rotated)

    # Random brightness/contrast
    for _ in range(2):
        alpha = random.uniform(0.8, 1.2)
        beta = random.randint(-30, 30)
        bright_contrast = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
        augmented.append(bright_contrast)

    # Small scaling + translation
    for _ in range(2):
        rows, cols, _ = img.shape
        scale = random.uniform(0.9, 1.1)
        tx = random.randint(-5, 5)
        ty = random.randint(-5, 5)
        M = np.float32([[scale, 0, tx], [0, scale, ty]])
        transformed = cv2.warpAffine(img, M, (cols, rows), borderMode=cv2.BORDER_REFLECT)
        augmented.append(transformed)

    random.shuffle(augmented)
    return augmented

# =========================
# Process folder
# =========================
def process_folder(input_folder, original_output):
    os.makedirs(original_output, exist_ok=True)
    processed_images = []
    
    for img_name in tqdm(os.listdir(input_folder), desc=f"Processing {os.path.basename(input_folder)}"):
        img_path = os.path.join(input_folder, img_name)
        try:
            img = cv2.imread(img_path)
            if img is None:
                continue

            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

            if len(faces) == 0:
                face = cv2.resize(img, target_size)
            else:
                x, y, w, h = sorted(faces, key=lambda f: f[2]*f[3], reverse=True)[0]
                face = img[y:y+h, x:x+w]
                face = cv2.resize(face, target_size)

            # CLAHE
            face = apply_clahe(face)

            # Save original
            out_path = os.path.join(original_output, img_name)
            cv2.imwrite(out_path, face)
            processed_images.append(img_name)

        except Exception as e:
            print(f"Error processing {img_name}: {e}")
    
    return processed_images

# =========================
# Preprocessing + Exact Augmentation
# =========================
def preprocess_all(input_root, output_root):
    original_root = os.path.join(output_root, "original")
    augmented_root = os.path.join(output_root, "augmented")
    os.makedirs(original_root, exist_ok=True)
    os.makedirs(augmented_root, exist_ok=True)

    class_counts = {}

    # Step 1: Process original images
    for class_name in os.listdir(input_root):
        input_class_folder = os.path.join(input_root, class_name)
        if os.path.isdir(input_class_folder):
            orig_class_folder = os.path.join(original_root, class_name)
            processed_images = process_folder(input_class_folder, orig_class_folder)
            class_counts[class_name] = len(processed_images)

    # Step 2: Smart augmentation for exact balance
    max_count = max(class_counts.values())
    for class_name, count in class_counts.items():
        if count < max_count:
            orig_class_folder = os.path.join(original_root, class_name)
            aug_class_folder = os.path.join(augmented_root, class_name)
            os.makedirs(aug_class_folder, exist_ok=True)

            images = os.listdir(orig_class_folder)
            needed = max_count - count

            while needed > 0:
                img_name = random.choice(images)
                img = cv2.imread(os.path.join(orig_class_folder, img_name))
                aug_images = augment_image(img)

                for aug_img in aug_images:
                    if needed <= 0:
                        break
                    new_name = f"aug_{uuid.uuid4().hex}_{img_name}"

                    cv2.imwrite(os.path.join(aug_class_folder, new_name), aug_img)
                    needed -= 1

    # Step 3: Summary (handles missing augmented folders)
    print("✅ Preprocessing complete! Class-wise image counts:")
    for class_name in os.listdir(original_root):
        orig_count = len(os.listdir(os.path.join(original_root, class_name)))
        aug_class_path = os.path.join(augmented_root, class_name)
        aug_count = len(os.listdir(aug_class_path)) if os.path.exists(aug_class_path) else 0
        total = orig_count + aug_count
        print(f"{class_name}: original={orig_count}, augmented={aug_count}, total={total}")

# =========================
# Merge originals + augmented
# =========================
def merge_original_augmented(original_root, augmented_root, merged_root):
    os.makedirs(merged_root, exist_ok=True)
    class_names = os.listdir(original_root)
    for class_name in tqdm(class_names, desc="Merging classes"):
        merged_class_folder = os.path.join(merged_root, class_name)
        os.makedirs(merged_class_folder, exist_ok=True)

        # Copy original
        orig_class_folder = os.path.join(original_root, class_name)
        for img_name in os.listdir(orig_class_folder):
            shutil.copy2(os.path.join(orig_class_folder, img_name), merged_class_folder)

        # Copy augmented if exists
        aug_class_folder = os.path.join(augmented_root, class_name)
        if os.path.exists(aug_class_folder):
            for img_name in os.listdir(aug_class_folder):
                shutil.copy2(os.path.join(aug_class_folder, img_name), merged_class_folder)

    print(f"✅ Merge complete! All images saved in {merged_root}")

# =========================
# Visualize samples before/after augmentation
# =========================
def visualize_samples(before_root, after_root, n=5):
    class_names = [c for c in os.listdir(before_root) if os.path.isdir(os.path.join(before_root, c))]
    chosen_class = random.choice(class_names)

    before_class_path = os.path.join(before_root, chosen_class)
    after_class_path = os.path.join(after_root, chosen_class)

    before_images = random.sample(os.listdir(before_class_path), min(n, len(os.listdir(before_class_path))))
    plt.figure(figsize=(12, 4))

    for i, img_name in enumerate(before_images):
        before_img = cv2.imread(os.path.join(before_class_path, img_name))
        after_img = cv2.imread(os.path.join(after_class_path, img_name))

        if before_img is not None and after_img is not None:
            before_img = cv2.cvtColor(before_img, cv2.COLOR_BGR2RGB)
            after_img = cv2.cvtColor(after_img, cv2.COLOR_BGR2RGB)

            plt.subplot(2, n, i+1)
            plt.imshow(before_img)
            plt.title(f"Before: {chosen_class}")
            plt.axis("off")

            plt.subplot(2, n, n+i+1)
            plt.imshow(after_img)
            plt.title(f"After: {chosen_class}")
            plt.axis("off")

    plt.tight_layout()
    plt.show()

# =========================
# Visualize merged dataset samples
# =========================
def visualize_merged_samples(merged_root, n=5):
    class_names = [c for c in os.listdir(merged_root) if os.path.isdir(os.path.join(merged_root, c))]
    chosen_class = random.choice(class_names)

    class_path = os.path.join(merged_root, chosen_class)
    images = random.sample(os.listdir(class_path), min(n, len(os.listdir(class_path))))

    plt.figure(figsize=(12, 4))
    for i, img_name in enumerate(images):
        img = cv2.imread(os.path.join(class_path, img_name))
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.subplot(1, n, i+1)
            plt.imshow(img)
            plt.title(f"{chosen_class}")
            plt.axis("off")

    plt.tight_layout()
    plt.show()

# =========================
# MAIN EXECUTION
# =========================
if __name__ == "__main__":
    # Clear previous output folders for a clean run
    clear_folder("images/dataset/train_cleaned/original")
    clear_folder("images/dataset/train_cleaned/augmented")
    clear_folder("images/dataset/train_cleaned/merged")

    # Step 1-2: Preprocess + exact augmentation
    preprocess_all(input_root, output_root)

    # Step 3: Visualize before/after augmentation
    visualize_samples(
        before_root=os.path.join(output_root, "original"),
        after_root=os.path.join(output_root, "augmented"),
        n=5
    )

    # Step 4: Merge original + augmented
    original_root = os.path.join(output_root, "original")
    augmented_root = os.path.join(output_root, "augmented")
    merged_root = os.path.join(output_root, "merged")

    merge_original_augmented(original_root, augmented_root, merged_root)

    # Step 5: Visualize final merged dataset
    visualize_merged_samples(merged_root, n=5)

```

```python
# =========================
# Final summary for merged dataset
# =========================
def print_merged_summary(merged_root):
    print("✅ Final merged dataset class-wise counts:")
    for class_name in os.listdir(merged_root):
        class_path = os.path.join(merged_root, class_name)
        if os.path.isdir(class_path):
            count = len(os.listdir(class_path))
            print(f"{class_name}: {count} images")
print_merged_summary(merged_root)

```

## ready-to-use PyTorch  data loader code for your merged dataset, fully compatible with CNN+LSTM trainin

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os

# =========================
# Custom Dataset
# =========================
class EmotionDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.images = []
        self.labels = []
        self.class_to_idx = {}
        for idx, class_name in enumerate(sorted(os.listdir(root_dir))):
            class_path = os.path.join(root_dir, class_name)
            if os.path.isdir(class_path):
                self.class_to_idx[class_name] = idx
                for img_name in os.listdir(class_path):
                    self.images.append(os.path.join(class_path, img_name))
                    self.labels.append(idx)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        label = self.labels[idx]
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label

# =========================
# Transformations
# =========================
transform = transforms.Compose([
    transforms.Resize((128, 128)),   # Ensure size matches your model input
    transforms.ToTensor(),           # Convert to PyTorch tensor
    transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5])  # Normalize to [-1,1]
])

# =========================
# Create Dataset and DataLoader
# =========================
merged_root = "images/dataset/train_cleaned/merged"
dataset = EmotionDataset(merged_root, transform=transform)

batch_size = 32
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=0)


# =========================
# Example iteration
# =========================
for images, labels in dataloader:
    print(images.shape, labels.shape)
    break
```

```python
class EmotionCNN(nn.Module):
    def __init__(self, num_classes=7):
        super(EmotionCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3,32,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32,64,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64,128,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128*16*16,256),
            nn.ReLU(),
            nn.Linear(256,num_classes)
        )

    def forward(self,x):
        x = self.features(x)
        x = self.classifier(x)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = EmotionCNN(num_classes=num_classes).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
num_epochs = 1  # start smaller for testing

batch_size = 16  # smaller batch for CPU
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=0)

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for batch_idx, (images, labels) in enumerate(dataloader):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        if batch_idx % 50 == 0:
            print(f"Epoch {epoch+1}/{num_epochs} | Batch {batch_idx}/{len(dataloader)} | Loss: {loss.item():.4f}")

    epoch_loss = running_loss / len(dataloader)
    epoch_acc = 100 * correct / total
    print(f"✅ Epoch {epoch+1} completed | Avg Loss: {epoch_loss:.4f} | Accuracy: {epoch_acc:.2f}%\n")

torch.save(model.state_dict(), "emotion_cnn.pth")
print("✅ Model trained and saved!")
```

```python
import cv2
import torch
from torchvision import transforms
from PIL import Image
from collections import defaultdict

# =========================
# Setup
# =========================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_classes = 7

# Load trained model
model = EmotionCNN(num_classes=num_classes).to(device)
model.load_state_dict(torch.load("emotion_cnn.pth", map_location=device))
model.eval()

# Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Transform
transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor(),
    transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5])
])

# Track faces
face_track = {}
face_id = 0

# =========================
# Mapping
# =========================
# Map emotions to Interested / Not Interested
emotion_map = {
    'happy': 'Interested',
    'surprise': 'Interested',
    'neutral': 'Interested',
    'sad': 'Not Interested',
    'angry': 'Not Interested',
    'fear': 'Not Interested',
    'disgust': 'Not Interested'
}

# idx -> emotion from dataset
idx_to_emotion = {v:k for k,v in dataset.class_to_idx.items()}
idx_to_interest = {idx: emotion_map[idx_to_emotion[idx]] for idx in idx_to_emotion.keys()}

# =========================
# Live Webcam Detection
# =========================
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        face_img = frame[y:y+h, x:x+w]
        face_pil = Image.fromarray(cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB))
        face_tensor = transform(face_pil).unsqueeze(0).to(device)

        # Predict
        with torch.no_grad():
            output = model(face_tensor)
            _, pred = output.max(1)
            pred = pred.item()
            emotion_label = idx_to_emotion[pred]
            interest_label = idx_to_interest[pred]

        # Assign temporary ID
        temp_id = face_id
        face_id += 1
        face_track[temp_id] = interest_label

        # Draw rectangle + labels
        color = (0,255,0) if interest_label=="Interested" else (0,0,255)
        cv2.rectangle(frame, (x,y), (x+w, y+h), color, 2)
        cv2.putText(frame, f"{emotion_label} | {interest_label}", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    cv2.imshow("Live Emotion Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# =========================
# Final Overall Percentage
# =========================
from collections import Counter
counts = Counter(face_track.values())
total_people = sum(counts.values())
if total_people > 0:
    interested_pct = counts['Interested'] / total_people * 100
    not_interested_pct = counts['Not Interested'] / total_people * 100
    print(f"✅ Final Overall: Interested = {interested_pct:.1f}%, Not Interested = {not_interested_pct:.1f}%")

cap.release()
cv2.destroyAllWindows()
```
